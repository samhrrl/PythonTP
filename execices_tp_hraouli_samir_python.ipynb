```python
L = [17,38,10,15,72,12]
L2 = [108,78,98,17,0,-8,4] #liste de test pour le trie
#trier la liste pour tout élement. On utilisera le trie par insertion qui est efficace pour les petites listes tel que la notre avec une complexité 0(n) = n²

r = "affichage de la liste {}".format(L)
print(r)

def tri_par_insertion(l):
    
    for i in range(1,len(l)): #parcours la longeur de la liste
        temp = l[i] #on stock l'élement i dans une variable temporaire
        j=i
        while j>0 and temp<l[j-1]:
            l[j]=l[j-1]
            j-=1
        l[j] = temp
    return l

r1 = "affichage de la liste avec les élements triés grace au tri par insertion : {}".format(tri_par_insertion(L2))
r2 = "affichage de la liste avec les élements triés grace au tri par insertion et l'ajout de l'élement 12 à la fin de la liste : {}".format(tri_par_insertion(L))
print(r1)
print(r2)
#print(tri_par_insertion(L2))
#print(tri_par_insertion(L)) #on voit bien que c'est trié lorsqu'on ajoute 12 à la liste

def inversion_liste(l):
    longueur = len(l)
    s = longueur
    for i in l:
        s = s-1
        l[s] = i
    return l

r3 = "affichage liste inversé: {} ".format(inversion_liste(L))
print(r3)

#affichage de l'indice de l'élement 17

#x est l'élement
#l est la liste

liste_test = [2,8,7,9,1,0]

def affichage_indice_element_liste(liste,x):
    
    for e in range(0, len(liste)): 
        
        element_liste = liste[e]
        
        if x == element_liste:
            return e

print(affichage_indice_element_liste(liste_test,1))

#A noter: si on teste pour les listes L et L2, ne pas oublier que les indices ont changé à cause du trie par insertion


def suppression_element_liste(l, e):
    
    for k in range(0, len(l)):
        if l[k] == e:
            liste = l.remove(e)
            return liste
            
        
suppression_element_liste(liste_test, 9)

#afficher sous liste du deuxième au troisième élement

t1 = "affichage liste du deuxième au troisième élement: {}".format(liste_test[1:3])
print(t1)

#afficher sous liste du début au deuxième élement

t2 = "affichage liste du premier au deuxième élement: {}".format(liste_test[:2])
print(t2)

#afficher sous liste du troisième élement au dernier élement

t3 = "affichage liste du troisième au dernier élement: {}".format(liste_test[2:])
print(t3)

#afficher dernier élement indicage négatif

t4 = "affichage liste du dernier élement: {}".format(liste_test[-1:])
print(t4)
```

```python
## Construire une liste des jours de la semaine correspondant aux dates dans la liste dates1 et afficher la nouvelle liste

from datetime import datetime

dates1 = ['05-10-2020', '06-10-2020', '07-10-2020', '08-10-2020', '09-10-2020']

# step by step 

#05-10-2020
date_1 = datetime.strptime(dates1[0], '%d-%m-%Y') #validé
print(date_1)
jour_semaine = date_1.strftime('%A')
print(jour_semaine)

#06-10-2020
date_2 = datetime.strptime(dates1[1], '%d-%m-%Y') #validé
print(date_2)
jour_semaine2 = date_2.strftime('%A')
print(jour_semaine2)

#07-10-2020
date_3 = datetime.strptime(dates1[2], '%d-%m-%Y') #validé
print(date_3)
jour_semaine3 = date_3.strftime('%A')
print(jour_semaine3)

#08-10-2020
date_4 = datetime.strptime(dates1[3], '%d-%m-%Y') #validé
print(date_4)
jour_semaine4 = date_4.strftime('%A')
print(jour_semaine4)

#09-10-2020
date_5 = datetime.strptime(dates1[4], '%d-%m-%Y') #validé
print(date_5)
jour_semaine5 = date_5.strftime('%A')
print(jour_semaine5)


# en assemblant

def affichage_jour_semaine_date(l):
    x = []
    for i in range(0, len(l)):
        date = datetime.strptime(l[i], '%d-%m-%Y')
        jour = date.strftime('%A')
        x.append(jour) #ajout des jour dans ma nouvelle liste
    return x
        
affichage_jour_semaine_date(dates1)
```
